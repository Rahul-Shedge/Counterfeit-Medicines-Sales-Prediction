# RANDOM FOREST 

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
Rgs = RandomForestRegressor()
```

```python
param_dist = {"n_estimators":[700,600,900,1000],
              "max_features": [5,9,7],
              "bootstrap": [True, False],
                'criterion':['mse','mae'],
                'max_depth':[None,5,7],
                'min_samples_leaf':[5,8], 
                'min_samples_split':[5,7,8,9]
                  }
```

```python
# run randomized search
n_iter_search = 10
# n_iter parameter of RandomizedSeacrhCV controls, how many 
# parameter combination will be tried; out of all possible given values

random_search = RandomizedSearchCV(Rgs, param_distributions=param_dist,
                                   n_iter=n_iter_search,scoring='neg_mean_squared_error',cv=5)
random_search.fit(x_train, y_train)
```

```python
random_search.best_estimator_
```

```python
rg=random_search.best_estimator_
```

```python
rg.fit(x_train,y_train)
```

```python
pred=rg.predict(x_test)
```

```python
MAE=mean_absolute_error(y_test,pred)
```
